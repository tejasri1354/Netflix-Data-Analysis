```python
import pandas as pd
df=pd.read_csv(r"C:\Users\shini\Downloads\netflix\netflix_titles.csv")
df
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df[df['director'].isnull()].head()
```

```python
df['director'].fillna('Unknown', inplace=True)
df.head()
```

```python
df[df['cast'].isnull()].head()
```

```python
df['cast']=df['cast'].fillna('Unknown')
df.head()
```

```python
df[df['country'].isnull()].head()
```

```python
df['country']=df['country'].fillna('unknown')
df.head()
```

```python
df.isnull().sum()
```

```python
df['date_added'].isnull().sum()
```

```python
df['date_added'].head(10)
```

```python
df['date_added'] = pd.to_datetime(
    df['date_added'],
    errors='coerce'
)
```

```python
df['date_added'].isnull().sum()
```

```python
df['added_year'] = df['date_added'].dt.year
```

```python
df
```

```python
df.isnull().sum()
    
```

```python
df['date_added'].fillna('0', inplace=True)
```

```python
df['added_year'].fillna('0', inplace=True)
```

```python
df['rating'].fillna('0', inplace=True)
```

```python
df['duration'].fillna('unknown', inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df['type'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(6,4))

sns.countplot(data=df, x='type',color='pink')

plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Type")
plt.ylabel("Count")

plt.show()
```

```python
df['country'].value_counts().head(10)
```

```python
plt.figure(figsize=(10,5))

df['country'].value_counts().head(10).plot(kind='bar',color='green')

plt.title("Top 10 Countries on Netflix")
plt.xlabel("Country")
plt.ylabel("Number of Titles")

plt.xticks(rotation=45)

plt.show()
```

```python
df["listed_in"].value_counts().head(10).plot(kind="bar")

plt.title("Top Netflix Genres")
plt.show()
```

```python
df["rating"].value_counts().head(10).plot(kind="bar")

plt.title("Netflix Ratings Distribution")

plt.show()
```
